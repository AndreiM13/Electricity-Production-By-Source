```python
# Assuming your dataframe is named 'df' and has a column named 'country' and a column named 'year'

# Get a list of unique countries in the dataframe
countries = df2['Country'].unique()

# Create an empty list to store the countries with missing data
missing_countries = []

# Loop through each country in the 'countries' list
for country in countries:
    # Get a subset of the dataframe for the current country
    country_df = df[df['Country'] == country]
    # Get a list of unique years in the subset dataframe
    years = country_df['Year'].unique()
    # Check if the number of years in the subset dataframe is equal to the number of years from the minimum year to the maximum year in the subset dataframe
    if len(years) != max(years) - min(years) + 1:
        # If the number of years is not equal, add the country to the 'missing_countries' list
        missing_countries.append(country)

# Print the list of countries with missing data
print("Countries with missing data:", missing_countries)
```
